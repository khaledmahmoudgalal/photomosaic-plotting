```python
# Importing libs :

import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# reading Imgs :

image = cv2.imread('sora1.jpg')

back = cv2.imread('background.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

edged = cv2.Canny(gray, 30, 200)

# refer to open cv documintaion ( Morphological Transformations) page . 68 :

kernel = np.ones((3,3),np.uint8)
edged = cv2.morphologyEx(edged, cv2.MORPH_CLOSE, kernel)

contours, hierarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

print("Number of Contours found = " + str(len(contours)))

# sorting the contours according to area :

contours.sort(key= lambda x:cv2.contourArea(x))

# detecting the bounding points to the contour with maximum area :

rect = cv2.minAreaRect(contours[0])
box = cv2.boxPoints(rect)

# masking :

h , w ,_=image.shape
mask = np.zeros((h,w,1))
mask = cv2.rectangle(mask,(box[1][0],box[1][1]),(box[3][0],box[3][1]),255,-1).astype("uint8")

print("box[1][0]=" + str(box[1][0]))
print("box[1][1]=" + str(box[1][1]))
print("box[3][0]=" + str(box[3][0]))
print("box[3][1]=" + str(box[3][1]))

image = cv2.bitwise_and(image,image,mask=mask)

# cropping :

starty = int(box[1][1])
endy = int(box[3][1])
startx = int(box[1][0])
endx = int(box[3][0])
cropped = image[starty:endy, startx:endx]

# resizing :

print('Original Dimensions : ',cropped.shape)

scale_percent = 30 # percent of original size
width = int(cropped.shape[1] * scale_percent / 100)
height = int(cropped.shape[0] * scale_percent / 100)
dim = (width, height)

resized = cv2.resize(cropped, dim, interpolation = cv2.INTER_AREA)

print('Resized Dimensions : ',resized.shape)

# background final output :

back[0:height, 0:width] = resized

# Images showing :

#cv2.imshow('Canny Edges After Contouring', edged)
#cv2.imshow('box',mask)
#cv2.imshow('image after mask', image)
#cv2.imshow('yarab', cropped)
#cv2.imshow("Resized image", resized)
#cv2.imshow('final',back)


#cv2.waitKey(0)
#cv2.destroyAllWindows()

#------------------------------------- (2) --------------------------------------------------------------------

# reading Imgs part 2 :

image2 = cv2.imread('sora2.jpg')

gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

edged2 = cv2.Canny(gray2, 30, 200)

# refer to open cv documintaion ( Morphological Transformations) page . 68 part 2 :

kernel2 = np.ones((3,3),np.uint8)
edged2 = cv2.morphologyEx(edged2, cv2.MORPH_CLOSE, kernel2)

contours2, hierarchy2 = cv2.findContours(edged2, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

print("Number of Contours found (2) = " + str(len(contours2)))

# sorting the contours according to area (2) :

contours2.sort(key= lambda x2:cv2.contourArea(x2))

# detecting the bounding points to the contour with maximum area (2) :

rect2 = cv2.minAreaRect(contours2[0])
box2 = cv2.boxPoints(rect2)

# masking (2) :

h2 , w2 ,_=image2.shape
mask2 = np.zeros((h2,w2,1))
mask2 = cv2.rectangle(mask2,(box2[1][0],box2[1][1]),(box2[3][0],box2[3][1]),255,-1).astype("uint8")

print("box2[1][0]=" + str(box2[1][0]))
print("box2[1][1]=" + str(box2[1][1]))
print("box2[3][0]=" + str(box2[3][0]))
print("box2[3][1]=" + str(box2[3][1]))

image2 = cv2.bitwise_and(image2,image2,mask=mask2)

# cropping (2) :

starty2 = int(box2[1][1])
endy2 = int(box2[3][1])
startx2 = int(box2[1][0])
endx2 = int(box2[3][0])
cropped2 = image2[starty2:endy2, startx2:endx2]

# resizing (2) :

print('Original Dimensions (2) : ',cropped2.shape)

scale_percent2 = 30 # percent of original size
width2 = int(cropped2.shape[1] * scale_percent2 / 100)
height2 = int(cropped2.shape[0] * scale_percent2 / 100)
dim2 = (width2, height2)

resized2 = cv2.resize(cropped2, dim2, interpolation = cv2.INTER_AREA)

print('Resized Dimensions(2): ',resized2.shape)

# background final output(2) :

back[height:(height+height2), 0:width2] = resized2

# Images showing (2) :

#cv2.imshow('Canny Edges After Contouring (2)', edged2)
#cv2.imshow('box (2)',mask2)
#cv2.imshow('image after mask (2)', image2)
#cv2.imshow('yarab (2)', cropped2)
#cv2.imshow("Resized image (2)", resized2)
#cv2.imshow('final (2)',back)

#cv2.waitKey(0)
#cv2.destroyAllWindows()

#------------------------------------- (3) --------------------------------------------------------------------

# reading Imgs part 3 :

image3 = cv2.imread('sora3.jpg')

gray3 = cv2.cvtColor(image3, cv2.COLOR_BGR2GRAY)

edged3 = cv2.Canny(gray3, 30, 200)

# refer to open cv documintaion ( Morphological Transformations) page . 68 part 3 :

kernel3 = np.ones((3,3),np.uint8)
edged3 = cv2.morphologyEx(edged3, cv2.MORPH_CLOSE, kernel3)

contours3, hierarchy3 = cv2.findContours(edged3, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

print("Number of Contours found (3) = " + str(len(contours3)))

# sorting the contours according to area (3) :

contours3.sort(key= lambda x3:cv2.contourArea(x3))

# detecting the bounding points to the contour with maximum area (3) :

rect3 = cv2.minAreaRect(contours3[0])
box3 = cv2.boxPoints(rect3)

# masking (3) :

h3 , w3 ,_=image3.shape
mask3 = np.zeros((h3,w3,1))
mask3 = cv2.rectangle(mask3,(box3[1][0],box3[1][1]),(box3[3][0],box3[3][1]),255,-1).astype("uint8")

print("box3[1][0]=" + str(box3[1][0]))
print("box3[1][1]=" + str(box3[1][1]))
print("box3[3][0]=" + str(box3[3][0]))
print("box3[3][1]=" + str(box3[3][1]))

image3 = cv2.bitwise_and(image3,image3,mask=mask3)

# cropping (3) :

starty3 = int(box3[1][1])
endy3 = int(box3[3][1])
startx3 = int(box3[1][0])
endx3 = int(box3[3][0])
cropped3 = image3[starty3:endy3, startx3:endx3]

# resizing (3) :

print('Original Dimensions (3) : ',cropped3.shape)

scale_percent3 = 30 # percent of original size
width3 = int(cropped3.shape[1] * scale_percent3 / 100)
height3 = int(cropped3.shape[0] * scale_percent3 / 100)
dim3 = (width3, height3)

resized3 = cv2.resize(cropped3, dim3, interpolation = cv2.INTER_AREA)

print('Resized Dimensions(3): ',resized3.shape)

# background final output(3) :

back[height:(height+height3), width2:(width2+width3)] = resized3

# Images showing (3) :

#cv2.imshow('Canny Edges After Contouring (3)', edged3)
#cv2.imshow('box (3)',mask3)
#cv2.imshow('image after mask (3)', image3)
#cv2.imshow('yarab (3)', cropped3)
#cv2.imshow("Resized image (3)", resized3)
#cv2.imshow('final (3)',back)

#cv2.waitKey(0)
#cv2.destroyAllWindows()

#------------------------------------- (4) --------------------------------------------------------------------

# reading Imgs part 4 :

image4 = cv2.imread('sora4.jpg')

gray4 = cv2.cvtColor(image4, cv2.COLOR_BGR2GRAY)

edged4 = cv2.Canny(gray4, 30, 200)

# refer to open cv documintaion ( Morphological Transformations) page . 68 part 4 :

kernel4 = np.ones((3,3),np.uint8)
edged4 = cv2.morphologyEx(edged4, cv2.MORPH_CLOSE, kernel4)

contours4, hierarchy4 = cv2.findContours(edged4, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

print("Number of Contours found (4) = " + str(len(contours4)))

# sorting the contours according to area (4) :

contours4.sort(key= lambda x4:cv2.contourArea(x4))

# detecting the bounding points to the contour with maximum area (4) :

rect4 = cv2.minAreaRect(contours4[0])
box4 = cv2.boxPoints(rect4)

# masking (4) :

h4 , w4 ,_=image4.shape
mask4 = np.zeros((h4,w4,1))
mask4 = cv2.rectangle(mask4,(box4[1][0],box4[1][1]),(box4[3][0],box4[3][1]),255,-1).astype("uint8")

print("box4[1][0]=" + str(box4[1][0]))
print("box4[1][1]=" + str(box4[1][1]))
print("box4[3][0]=" + str(box4[3][0]))
print("box4[3][1]=" + str(box4[3][1]))

image4 = cv2.bitwise_and(image4,image4,mask=mask4)

# cropping (4) :

starty4 = int(box4[1][1])
endy4 = int(box4[3][1])
startx4 = int(box4[1][0])
endx4 = int(box4[3][0])
cropped4 = image4[starty4:endy4, startx4:endx4]

# resizing (4) :

print('Original Dimensions (4) : ',cropped4.shape)

scale_percent4 = 30 # percent of original size
width4 = int(cropped4.shape[1] * scale_percent4 / 100)
height4 = int(cropped4.shape[0] * scale_percent4 / 100)
dim4 = (width4, height4)

resized4 = cv2.resize(cropped4, dim4, interpolation = cv2.INTER_AREA)

print('Resized Dimensions(4): ',resized4.shape)

# background final output(4) :

back[height:(height+height4), (width2+width3):(width2+width3+width4)] = resized4

# Images showing (4) :

#cv2.imshow('Canny Edges After Contouring (4)', edged4)
#cv2.imshow('box (4)',mask4)
#cv2.imshow('image after mask (4)', image4)
#cv2.imshow('yarab (4)', cropped4)
#cv2.imshow("Resized image (4)", resized4)
#cv2.imshow('final (4)',back)

#cv2.waitKey(0)
#cv2.destroyAllWindows()

#------------------------------------- (5) --------------------------------------------------------------------

# reading Imgs part 5 :

image5 = cv2.imread('sora5.jpg')

gray5 = cv2.cvtColor(image5, cv2.COLOR_BGR2GRAY)

edged5 = cv2.Canny(gray5, 30, 200)

# refer to open cv documintaion ( Morphological Transformations) page . 68 part 5 :

kernel5 = np.ones((3,3),np.uint8)
edged5 = cv2.morphologyEx(edged5, cv2.MORPH_CLOSE, kernel5)

contours5, hierarchy5 = cv2.findContours(edged5, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

print("Number of Contours found (5) = " + str(len(contours5)))

# sorting the contours according to area (5) :

contours5.sort(key= lambda x5:cv2.contourArea(x5))

# detecting the bounding points to the contour with maximum area (5) :

rect5 = cv2.minAreaRect(contours5[0])
box5 = cv2.boxPoints(rect5)

# masking (5) :

h5 , w5 ,_=image5.shape
mask5 = np.zeros((h5,w5,1))
mask5 = cv2.rectangle(mask5,(box5[1][0],box5[1][1]),(box5[3][0],box5[3][1]),255,-1).astype("uint8")

print("box5[1][0]=" + str(box5[1][0]))
print("box5[1][1]=" + str(box5[1][1]))
print("box5[3][0]=" + str(box5[3][0]))
print("box5[3][1]=" + str(box5[3][1]))

image5 = cv2.bitwise_and(image5,image5,mask=mask5)

# cropping (5) :

starty5 = int(box5[1][1])
endy5 = int(box5[3][1])
startx5 = int(box5[1][0])
endx5 = int(box5[3][0])
cropped5 = image5[starty5:endy5, startx5:endx5]

# resizing (5) :

print('Original Dimensions (5) : ',cropped5.shape)

scale_percent5 = 30 # percent of original size
width5 = int(cropped5.shape[1] * scale_percent5 / 100)
height5 = int(cropped5.shape[0] * scale_percent5 / 100)
dim5 = (width5, height5)

resized5 = cv2.resize(cropped5, dim5, interpolation = cv2.INTER_AREA)

print('Resized Dimensions(5): ',resized5.shape)

# background final output(5) :

back[height:(height+height5), (width2+width3+width4):(width2+width3+width4+width5)] = resized5

# Images showing (5) :

#cv2.imshow('Canny Edges After Contouring (5)', edged5)
#cv2.imshow('box (5)',mask5)
#cv2.imshow('image after mask (5)', image5)
#cv2.imshow('yarab (5)', cropped5)
#cv2.imshow("Resized image (5)", resized5)
cv2.imshow('final (5)',back)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
